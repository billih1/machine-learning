```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

X, y = make_moons(n_samples=300, noise=0.05, random_state=42)

dbscan = DBSCAN(eps=0.2, min_samples=5)
y_pred = dbscan.fit_predict(X)

plt.figure(figsize=(8, 5))
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='rainbow')
plt.title("DBSCAN Clustering Result")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN

X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=42)

dbscan = DBSCAN(eps=0.6, min_samples=5)  # Adjust eps to match cluster density
y_pred = dbscan.fit_predict(X)

plt.figure(figsize=(8, 5))
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='rainbow')
plt.title("DBSCAN Clustering Result")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

X, y = make_circles(n_samples=500, factor=0.3, noise=0.05, random_state=42)
X = StandardScaler().fit_transform(X)

dbscan = DBSCAN(eps=0.2, min_samples=5)
y_pred = dbscan.fit_predict(X)

plt.figure(figsize=(7, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='rainbow', s=50)
plt.title("DBSCAN on Concentric Circles")
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_swiss_roll
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

X, _ = make_swiss_roll(n_samples=1000, noise=0.2)
X = PCA(n_components=2).fit_transform(X)
X = StandardScaler().fit_transform(X)

dbscan = DBSCAN(eps=0.3, min_samples=5)
y_pred = dbscan.fit_predict(X)

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='rainbow', s=15)
plt.title("DBSCAN on Swiss Roll (PCA 2D)")
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

X, y = make_blobs(n_samples=500, centers=3, cluster_std=1.0, random_state=42)
# Apply linear transformation to create anisotropy
transformation = [[0.6, -0.6], [-0.4, 0.8]]
X = X @ transformation
X = StandardScaler().fit_transform(X)

dbscan = DBSCAN(eps=0.3, min_samples=5)
y_pred = dbscan.fit_predict(X)

plt.figure(figsize=(7, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='rainbow', s=50)
plt.title("DBSCAN on Anisotropic Blobs")
plt.show()
```
