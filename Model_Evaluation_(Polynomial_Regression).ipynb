```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.datasets import load_diabetes

# Load dataset
data = load_diabetes()
X = data.data[:, 0].reshape(-1, 1)   # Use 1 feature so polynomial regression makes sense
y = data.target

# Polynomial features (degree=2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Train model
model = LinearRegression()
model.fit(X_poly, y)

# Predict
y_pred = model.predict(X_poly)

# Plot results
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(np.sort(X, axis=0),
         y_pred[np.argsort(X[:,0])],
         color='red',
         label='Polynomial Regression')

plt.xlabel('Feature 1 (example)')
plt.ylabel('Target')
plt.legend()
plt.show()
```

```python
from sklearn.datasets import fetch_california_housing
data = fetch_california_housing()
X = data.data[:, 0].reshape(-1, 1)  # Use first feature (MedInc)
y = data.target

# Polynomial regression
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, y)
y_pred = model.predict(X_poly)

plt.scatter(X, y, color='blue')
plt.plot(np.sort(X, axis=0), y_pred[np.argsort(X[:,0])], color='red')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.show()
```

```python
from sklearn.datasets import make_regression
X, y = make_regression(n_samples=100, n_features=1, noise=50, random_state=42)

poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, y)
y_pred = model.predict(X_poly)

plt.scatter(X, y, color='blue')
plt.plot(np.sort(X, axis=0), y_pred[np.argsort(X[:,0])], color='red')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.show()
```
