```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import pandas as pd
from sklearn.datasets import load_diabetes

diabetes = load_diabetes()
df = pd.DataFrame({
    'Feature': diabetes.data[:, 2],
    'Target': diabetes.target
})

X = df[['Feature']]
y = df['Target']

model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
residuals = y - y_pred

plt.scatter(X, residuals)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('BMI')
plt.ylabel('Residuals')
plt.title('Residual Plot - Diabetes (BMI)')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import pandas as pd
from sklearn.datasets import make_regression

X_data, y_data = make_regression(n_samples=100, n_features=1, noise=20, random_state=1)
df = pd.DataFrame({
    'Feature': X_data[:, 0],
    'Target': y_data
})

X = df[['Feature']]
y = df['Target']

model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
residuals = y - y_pred

plt.scatter(X, residuals)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Feature')
plt.ylabel('Residuals')
plt.title('Residual Plot - Synthetic Regression')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import pandas as pd
from sklearn.datasets import fetch_openml

wine = fetch_openml(name='wine-quality-red', as_frame=True)
df = pd.DataFrame({
    'Feature': wine.data['alcohol'].values,
    'Target': wine.target.astype(float)
})

X = df[['Feature']]
y = df['Target']

model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
residuals = y - y_pred

plt.scatter(X, residuals)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Alcohol')
plt.ylabel('Residuals')
plt.title('Residual Plot - Wine Quality (Alcohol)')
plt.show()
```
