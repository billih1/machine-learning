```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_diabetes

diabetes = load_diabetes()
df = pd.DataFrame({
    'Feature': diabetes.data[:, 2],
    'Target': diabetes.target
})

sns.scatterplot(x='Feature', y='Target', data=df)
plt.title("Diabetes Dataset - BMI vs Target")
plt.xlabel("BMI")
plt.ylabel("Target")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.datasets import fetch_california_housing

california = fetch_california_housing()
df = pd.DataFrame({
    'Feature': california.data[:, 3],
    'Target': california.target
})

sns.scatterplot(x='Feature', y='Target', data=df)
plt.title("California Housing - AveOccup vs Median House Value")
plt.xlabel("Average Occupants")
plt.ylabel("Median House Value")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.datasets import make_regression

X, y = make_regression(n_samples=100, n_features=1, noise=20, random_state=1)
df = pd.DataFrame({
    'Feature': X[:,0],
    'Target': y
})

sns.scatterplot(x='Feature', y='Target', data=df)
plt.title("Synthetic Regression Dataset")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.datasets import fetch_openml

wine = fetch_openml(name='wine-quality-red', as_frame=True)
df = pd.DataFrame({
    'Feature': wine.data['alcohol'].values,
    'Target': wine.target.astype(float)
})

sns.scatterplot(x='Feature', y='Target', data=df)
plt.title("Wine Quality - Alcohol vs Quality")
plt.xlabel("Alcohol")
plt.ylabel("Quality")
plt.show()
```
